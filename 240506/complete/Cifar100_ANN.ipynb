```python
import numpy as np
from keras.datasets import cifar100
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.utils import to_categorical
```

```python
(x_train_load, y_train_load), (x_test_load, y_test_load) = cifar100.load_data()
# 處理數據
x_train_load = x_train_load[:500]
y_train_load = y_train_load[:500]

x_test_load = x_test_load[:20]
y_test_load = y_test_load[:20]

## To check the class consists of the total class (eg.cifar100 has 100 classes )
full_set = set(range(100))
ytrain_set = set(y_train_load.flatten())     # !!! y_train_load 是一個包含三個列表的列表，因此透過.flatten()將它改成一個簡單的一維列表
# print(y_train_load)
missing_numbers = full_set - ytrain_set
print(len(missing_numbers))
print(missing_numbers)
```

```python
# Data normalization

# x_train = x_train_load.reshape((x_train_load.shape[0], 28, 28, 3)).astype('float32') / 255     # ==> 39s
# x_test = x_test_load.reshape((x_test_load.shape[0], 28, 28, 3)).astype('float32') / 255
x_train = x_train_load / 255     # ==> 39s
x_test = x_test_load / 255

# 將標籤轉換為 one-hot 編碼
# y_train = to_categorical(y_train_load)
# y_test = to_categorical(y_test_load)
y_train = (y_train_load)
y_test = (y_test_load)

# print(x_test.shape)
print(y_test)
```

```python
# 建立模型
model = Sequential([
    Flatten(input_shape=(x_train_load[0].shape)),
    Dense(20, activation='relu'),
    Dense(100, activation='softmax')
])

# 編譯模型
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',                       # < == # of class of train and test set are inconsisted :add 'sparse_'
              metrics=['accuracy'])
# 訓練模型
history = model.fit(x_train, y_train, epochs=50, batch_size=100, validation_data=(x_test, y_test))

```

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(x_train)
y_pred_label = np.argmax(y_pred, axis=1)
y_train_label = np.argmax(y_train, axis=1)
cm = confusion_matrix(y_train_label, y_pred_label)

# 打印混淆矩陣
print("Confusion Matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 繪製混淆矩陣的熱圖
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], 'r', label='Training loss')
plt.plot(history.history['val_loss'], 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
# 評估模型
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test a|ccuracy:', test_acc)
```

```python
y_pred = model.predict(x_test)
print(y_pred.shape)

y_pred_label = np.argmax(y_pred, axis=1)    
print('y_pred: ', y_pred_label)
```

```python
# 找出預測錯誤的索引
y_pred = model.predict(x_test)
print('y_shape: ', y_pred.shape)

y_pred_label = np.argmax(y_pred, axis=1)  # axis=1：沿著行查找每行的最大值的索引。axis=0：沿著列查找每列的最大值的索引。
y_test_label = np.argmax(y_test, axis=1)

print('y_pred: ', y_pred_label)
print('y_test: ', y_test_label)

len(x_test)
errors = np.where(y_pred_label != y_test_label)[0]
print(f'Index of Prediction errors:  {errors} ')
print(f'Total number of prediction errors: {len(errors)}')
```

```python
# 使用 matplotlib 顯示多張圖像
# 設置圖像數量

num_images = len(errors)

# 創建一個圖形和一組子圖
fig, axes = plt.subplots(1, num_images, figsize=(10, 10))

for i, ax in enumerate(axes):
    # 顯示第 i 張圖像
    ax.imshow(x_test_load[errors[i]])
    ax.set_title(f'Label: {y_pred_label[errors[i]]}')
    ax.axis('off')  # 不顯示軸

plt.tight_layout()
plt.show()
```
